```python
import face_recognition
import cv2
import numpy as np
import mediapipe as mp
import socket
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
```

```python
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
def login():
    global name
    cap = cv2.VideoCapture(0)

   
    nour_image = face_recognition.load_image_file("people/nour.jpg")
    nour_face_encoding = face_recognition.face_encodings(nour_image)[0]

    bassel_image = face_recognition.load_image_file("people/bassel.jpg")
    bassel_face_encoding = face_recognition.face_encodings(bassel_image)[0]
    
    abdallah_image = face_recognition.load_image_file("people/abdallah.jpg")
    abdallah_face_encoding = face_recognition.face_encodings(abdallah_image)[0]


    known_face_encodings = [
        nour_face_encoding,
        bassel_face_encoding,
        abdallah_face_encoding

    ]
    known_face_names = [
        "nour",
        "bassel",
        "abdallah"
        
    ]

    face_locations = []
    face_encodings = []
    face_names = []
    process_this_frame = True

    while True:
        ret, frame = cap.read()

        if process_this_frame:
            small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)
            face_locations = face_recognition.face_locations(small_frame)
            face_encodings = face_recognition.face_encodings(small_frame, face_locations)

            face_names = []
            for face_encoding in face_encodings:
                matches = face_recognition.compare_faces(known_face_encodings, face_encoding)
                name = "Unknown"
                face_distances = face_recognition.face_distance(known_face_encodings, face_encoding)
                best_match_index = np.argmin(face_distances)
                if matches[best_match_index]:
                    name = known_face_names[best_match_index]

                face_names.append(name)


        process_this_frame = not process_this_frame


        for (top, right, bottom, left), name in zip(face_locations, face_names):
            top *= 4
            right *= 4
            bottom *= 4
            left *= 4

            # Draw a box around the face
            cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)

            # Draw a label with a name below the face
            cv2.rectangle(frame, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
            font = cv2.FONT_HERSHEY_DUPLEX
            cv2.putText(frame, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)


        cv2.imshow('Video', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break


    cap.release()
    cv2.destroyAllWindows()

    

```

```python
import socket
import threading

def handle_client(conn, addr):
    print(f"Accepted connection from {addr}")
    

    login()
    
    msg = bytes(name, 'utf-8')
    conn.send(msg)
    
    conn.close()
    print(f"Connection from {addr} closed")

def start_server():
    host = 'localhost'
    port = 5000

    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind((host, port))
    server_socket.listen(5)

    print(f"Server listening on {host}:{port}")

    while True:
        try:
            conn, addr = server_socket.accept()
            client_thread = threading.Thread(target=handle_client, args=(conn, addr))
            client_thread.start()
        except Exception as e:
            print(f"Error accepting connection: {e}")

if __name__ == "__main__":
    start_server()
```
